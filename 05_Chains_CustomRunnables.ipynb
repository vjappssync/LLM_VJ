```python
import os
from dotenv import load_dotenv
load_dotenv()

if os.environ.get("GEMINI_API_KEY"):
    print("Api Key Exists")
    print(os.getenv("GEMINI_API_KEY"))
else:
    print("Api Key not exists")

load_dotenv(override=True)

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# TASK -1 [Prompt]
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])

# TASK - 2 [LLM]
llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# TASK - 3 [Str Parser]
str_parser = StrOutputParser()
```

```python
# TASK - 4 [Custom Runnable]
from langchain_core.runnables import RunnableLambda

def dictionary_maker(text:str) -> dict:
    return {"text": text}

dictionary_maker_runnableLambda = RunnableLambda(dictionary_maker)
```

```python
# TASK - 5 [Template For Post]

prompt_post = ChatPromptTemplate.from_messages(
    messages=[
        ("system", "You're a social media post generator. "),
        ("human", "Create a post for the following text for LinkedIn: {text}"),
    ]
)
```

```python
# TASK - 6 [Generate the Post using Gemini]

llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
# TASK - 7 [Str Parser]

str_parser = StrOutputParser()
```

```python
# FINALLY -> CHAIN

chain = prompt_template | llm_model | str_parser | dictionary_maker_runnableLambda | prompt_post | llm_model | str_parser

chain.invoke("What is the capital of tAIWAN?")
```
