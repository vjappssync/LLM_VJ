```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

message = [
    ("system","You are an helpful assistant"),
    ("human", "tell me about Vijay Vigneswar Selvakumar")
]
llm_model = ChatOpenAI(model="gpt-5.4-nano")
str_parser = StrOutputParser()
response = str_parser.parse(llm_model.invoke(message).content)
print(response)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/vj.txt")
loaded_data = loader.load()
loaded_data
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages(
    [
        ("human", "Answer this {question}, here is some extra {context}"),
    ]
)

messages = chat_template.format_messages(
    name="Vijay Vigneswar Selvakumar",
    question="Tell me about Vijay Vigneswar Selvakumar",
    context=loaded_data
)

response = llm_model.invoke(messages)
```
