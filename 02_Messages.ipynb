```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_openai import ChatOpenAI

llmmodel = ChatOpenAI(model="gpt-5.4-nano", temperature=1)

my_messages = [
    HumanMessage("Hey kid, tell a fun fact about US President Donald Trump"),
    SystemMessage("You are a Gen-Z Kid, response in a very funny way.")
]

llmmodel.invoke(my_messages).content
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

llmmodel = ChatOpenAI(model="gpt-5.4-nano", temperature=1)
user_input = input("Enter a topic for fun fact about : ")
# user_input = input("tell a fun fact about US President Donald Trump")
dynamic_prompt = PromptTemplate.from_template("tell a fun fact about {topic}")

llm_prompt = dynamic_prompt.invoke({"topic": user_input})

# print(llm_prompt)

llmmodel.invoke(llm_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_templates = ChatPromptTemplate.from_messages(
    [
        "Human", "tell a fun fact about {topic}",
        "System", "You are a Gen-Z Kid, always respond in {tone}"
    ]
)

user_topic = input("enter the topic: ")
system_tone = input("enter the tone: ")

ready_prompt = prompt_templates.invoke({"topic": user_topic, "tone": system_tone })

llmmodel.invoke(ready_prompt).content
```
