```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key = os.getenv("GEMINI_API_KEY"))
# llm_model = ChatOpenAI(model="gpt-5.4-nano", temperature=1)
result = llm_model.invoke("tell me a joke, output should be in key-value pair format with the following keys: setup, punchline.")

result.content
```

```python
# TASK -1 [Prompt]

prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant"),
    ("human", "{input}")
])
```

```python
# TASK - 2 [LLM]

llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
```

```python
# TASK - 3 [Str Parser]
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

```python
chain = prompt_template | llm_model | str_parser
chain.invoke({"input": "Capital of Tamilnadu?"})
```

```python
from langchain_core.Runnables import RunnableSequence

chain_rs = RunnableSequence(prompt_template, llm_model, str_parser)
chain.invoke({"input": "Capital of India?"})
```
