```python
from langchain_google_genai import ChatGoogleGenerativeAI
# from langchain_openai import ChatOpenAI

llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key = os.getenv("GEMINI_API_KEY"))
# llm_model = ChatOpenAI(model="gpt-5.4-nano", temperature=1)
result = llm_model.invoke("tell me a joke, output should be in key-value pair format with the following keys: setup, punchline.")

result.content
```

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    setup: str = Field(description="The setup for the joke")
    punchline: str = Field(description="The punchline for the joke")
```

```python
llm_structured_output = llm_model.with_structured_output(llm_schema)

result = llm_structured_output.invoke("tell me a joke")
print(result.setup)
print(result.punchline)
type(result)
```

```python
from typing import TypedDict

class llm_schema_td(TypedDict):
    setup: str
    punchline: str 
```

```python
llm_structured_output_td = llm_model.with_structured_output(llm_schema_td)

result = llm_structured_output_td.invoke("tell me a joke")
# print(result.setup) ==> THROWS ERROR
# print(result.punchline) ==> THROWS ERROR
result
# type(result) ==> DICT
```
