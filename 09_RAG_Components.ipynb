```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
# from langchain_chroma import Chroma
from langchain_community.vectorstores import FAISS


# Load the document, split it into chunks, embed each chunk and load it into the vector store.
loaded_document = TextLoader('./data/state_of_the_union.txt',
    encoding="utf-8").load()

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

chunks_of_text = text_splitter.split_documents(loaded_document)

```

```python
len(chunks_of_text)
```

```python
# vector_db = Chroma.from_documents(chunks_of_text, OpenAIEmbeddings())
vector_db = FAISS.from_documents(chunks_of_text, OpenAIEmbeddings())
```

```python
vector_db
```

```python
question = "What did the president say about the John Lewis Voting Rights Act?"

response = vector_db.similarity_search(question)

print(response[0].page_content)
```

```python
retriever = vector_db.as_retriever(search_kwargs={"k": 2})
```

```python
response = retriever.invoke("what did he say about ketanji brown jackson?")
response
```

```python
len(response)
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

template = """Answer the question based only on the following context:

{context}

Question: {question}
"""

prompt = ChatPromptTemplate.from_template(template)

model = ChatOpenAI()
```

LangChain LCEL pipeline used in a RAG flow

It takes a user question, retrieves relevant docs, formats them into context, sends everything to the prompt + model, and returns plain text.

```python
def format_docs(docs):
    return "\n\n".join([d.page_content for d in docs])
```

Assume that retriever object has below content:

[
  Document(page_content="Azure is cloud platform"),
  Document(page_content="Functions are serverless")
]

When you pass the retriever object to the method format_docs() you will get only the content as:

Azure is cloud platform
Functions are serverless

Now split this line 
{
    "context": retriever | format_docs,    STEP 2
    "question": RunnablePassthrough()      STEP 1
}

STEP 1: Using the RunnablePassthrough(), the "question" is set based on the query is passed while invoking the CHAIN

STEP 2: "Context" has retriever object and based on the question passed while invoking the CHAIN, it returns the docs based on the K value and it gets formatted using format_docs() method




Full Visual Flow
=================
User Question

   ↓
   
Create:

        context = retriever(question) → format_docs()

        question = same input

           ↓

Prompt Template

   ↓

LLM

   ↓

Plain Text Output


```python

chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
response = chain.invoke("what did he say about ketanji brown jackson?")
```
